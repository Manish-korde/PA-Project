### 1. Introduction to Generative Models
Generative models learn the underlying probability distribution of a dataset to generate new, synthetic data points. In this project, we utilize GANs for **Data Augmentation** and **Image Enhancement** (FaceAPP for Soil).

### 2. GAN Architecture
A GAN consists of two neural networks competing in a zero-sum game:
1. **Generator (G):** Takes random noise and tries to produce a realistic soil image.
2. **Discriminator (D):** Takes an image (real or synthetic) and tries to predict if it is 'Real' or 'Fake'.

### 3. Training Objective (BCE vs. Wasserstein Loss)
Standard GANs use **Binary Cross-Entropy (BCE) Loss**, but this often leads to **Mode Collapse** (where the generator only produces one type of image).

**Wasserstein GAN (WGAN)** solves this by using the **Earth Mover's Distance**:
- **W-Loss:** Measures the cost of transforming one distribution into another.
- **Lipschitz Continuity:** Enforced via weight clipping or gradient penalty to ensure stability.

### 4. Implementation Code (DCGAN Structure)


```python
import torch
import torch.nn as nn

# Generator Model (Unit 6: DCGAN)
class Generator(nn.Module):
    def __init__(self, z_dim, img_channels, feature_g):
        super(Generator, self).__init__()
        self.gen = nn.Sequential(
            # Input: N x z_dim x 1 x 1
            nn.ConvTranspose2d(z_dim, feature_g * 16, 4, 1, 0),
            nn.BatchNorm2d(feature_g * 16),
            nn.ReLU(),
            # Transposed convolutions to upsample image
            nn.ConvTranspose2d(feature_g * 16, feature_g * 8, 4, 2, 1),
            nn.BatchNorm2d(feature_g * 8),
            nn.ReLU(),
            nn.ConvTranspose2d(feature_g * 8, img_channels, 4, 2, 1),
            nn.Tanh() # Final output in range [-1, 1]
        )

    def forward(self, x):
        return self.gen(x)

print("DCGAN Generator Structure Loaded successfully for Unit 6 demo.")
```
